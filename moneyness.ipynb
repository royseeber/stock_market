## Calculate "Moneyness" for all strike prices on a given expiry date

```python
import pandas as pd
```

```python
#get a listing of objects in expiry-week_data bucket
from google.cloud import storage
client = storage.Client(project='Expiry Week')
bucket = client.get_bucket('expiry-week-data')

for item in bucket.list_blobs():
    print(item.name)
```

```python
#capture just 1 day's worth of data for development purposes
#pandas can read directly from a google cloud starage buck
df = pd.read_csv('gs://expiry-week-data/options/SPY_2020.csv')
df_sample = df[df['quotedate'] == '01/03/2020']
df_sample.to_csv('gs://expiry-week-data/options/sample_20200103.csv', index=False)
del df
```
